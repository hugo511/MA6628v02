# Brownian motion as a Vasicek

```python
x, y = vasicek(0., 0., 1., 1., 1., 2000)
plt.plot(x,y)
```

We compute $\mathbb E [W^2(1)]$ by simulation, which supposed to be $1.$. The result confirms the correction of the code in Euler scheme.

```python
ssum = 0
Total_N = 100
for i in range(Total_N):
    x, y = vasicek(0., 0., 1., 1., 1., 2000, seed = np.random.randint(0, 10000000))
    #print(y[-1])
    ssum = ssum + y[-1]**2
    
print('This is approximate second moment', ssum/Total_N)    
```
